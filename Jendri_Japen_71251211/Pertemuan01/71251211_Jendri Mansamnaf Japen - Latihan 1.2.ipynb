```python
emas1 = 25
harga1 = 650000
harga_emas1  = 685000

modal_awal = emas1 * harga1
nilai_pertama = emas1 * harga_emas1
untung_rp1 = nilai_pertama - modal_awal
untung_persen1 = (untung_rp1 / modal_awal) * 100

print(f"Keuntungang Awal : Rp {untung_rp1}")
print(f"Persentase Awal : {untung_persen1}%")

emas2 = 15
harga2 = 685000
harga_emas2 = 715000

modal_2 = emas2 * harga2
total_2 = modal_awal + modal_2
total_emas = emas1 + emas2
nilai = total_emas * harga_emas2
untung_rp2 = nilai - total_2
untung_persen2 = (untung_rp2 / total_2) * 100

print(f" Keuntungan 2 : Rp {untung_rp2}")
print(f"Persentase 2 : {untung_persen2}%")
```
