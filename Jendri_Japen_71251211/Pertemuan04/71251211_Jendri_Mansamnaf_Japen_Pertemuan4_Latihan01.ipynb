```python
#Latihan 4.1
#Demam atau Tidak
try:
    suhu = float(input("Masukan suhu tubuh: "))
    if suhu >= 38:
        print("Anda demam")
    else:
        print("Anda tidak demam")

except:
    print("Input tidak valid. Masukan angka suhu yang benar.")
```

```python
#Positif atau Negatif
try:
    bilangan = int(input("Masukan suatu bilangan: "))
    if bilangan >0:
        print("Positif")
    elif bilangan < 0:
        print("Negatif")
    else:
        print("Nol")
except:
    print("Input tidak valid. Masukan angka yang benar.")
```

```python
#Nilai Terbesar
try:
    x = int(input("Masukan bilangan pertama: "))
    y = int(input("Masukan bilangan kedua: "))
    z = int(input("Masukan bilangan ketiga: "))

    if x > y and x > z:
        print("Nilai terbesar:", x)
    elif y > x and y > z:
        print("Nilai terbesar:", y)
    else:
        print("Nilai terbesar:", z)
except:
    print("Input tidak valid. Semua input harus angka.")
```
