```python
#latihan 6.1
def perkalian(a, b):
    hasil = 0
    proses= ""

    for i in range(a):
        hasil += b
        proses += str(b)
        if i < a - 1:
            proses += " + "

    print(f"{a} x {b} = {proses} = {hasil}")

if __name__ == "__main__":
    a = int(input("Masukan angka pertama: "))
    b = int(input("Masukan angka kedua: "))

    perkalian(a, b)
```

```python
#latihan 6.2
def ganjil(bawah, atas):
    hasil = []

    if bawah < atas:
        for i in range(bawah, atas + 1):
            if i % 2 != 0:
                hasil.append(str(i))
    else:
        for i in range(bawah, atas - 1, -1):
            if i % 2 != 0:
                hasil.append(str(i))
    print(",".join(hasil))

if __name__ == "__main__":
    bawah = int(input("Masukan batas bawah: "))
    atas = int(input("Masukan batas atas: "))

    ganjil(bawah, atas)
```

```python
#latihan 6.3
def hitung_ips(jumlah_mk):
    total_bobot = 0

    for i in range(1, jumlah_mk + 1):
        nilai = input(f"Nilai MK {i}: ").upper()

        if nilai =="A":
            bobot = 4
        elif nilai =="B":
            bobot = 3
        elif nilai == "C":
            bobot = 2
        elif nilai == "D":
            bobot = 1
        else:
            bobot = 0

        total_bobot += bobot

    ips = total_bobot / jumlah_mk
    print("Nilai IPS anda semester ini %.2f" % ips)

if __name__ == "__main__":
    jumlah = int(input("Berapa jumlah mata kuliah? "))
    hitung_ips(jumlah)
```
