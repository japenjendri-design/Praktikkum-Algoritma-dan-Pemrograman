```python
#latihan 5.1
def cek_angka(a, b, c):
    if a != b and a != c and b != c:
        if a + b == c or a + c == b or b + c == a:
            return True
    return False
    
print(cek_angka(2, 3, 5))
print(cek_angka(4, 4, 8))
print(cek_angka(1, 2, 4))
```

```python
#latihan 5.2
def cek_digit_belakang(a, b, c):
    d1 = a % 10
    d2 = b % 10
    d3 = c % 10

    if d1 == d2 or d1 == d3 or d2 == d3:
        return True
    else:
        return False


a = int(input("Masukkan angka pertama: "))
b = int(input("Masukkan angka kedua: "))
c = int(input("Masukkan angka ketiga: "))

print(cek_digit_belakang(a, b, c))
```

```python
#latihan 5.3
celcius_to_fahrenheit = lambda c: (9/5) * c + 32
celcius_to_reamur = lambda c: 0.8 * c

print("Input C = 100 -> F =", celcius_to_fahrenheit(100))
print("Input C = 80 -> R =", celcius_to_reamur(80))
print("Input C = 0 -> F =", celcius_to_fahrenheit(0))
```
