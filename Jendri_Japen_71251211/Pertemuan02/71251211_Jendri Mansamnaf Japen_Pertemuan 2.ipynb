```python
#latihan 2.1

print("(==== KALKULATOR TARGET BERAT BADAN ====)")

tinggi = float(input("Masukan tinggi badan (m): "))
bmi = float(input("Masukan nilai BMI: "))
berat = bmi * (tinggi ** 2)
print("Berat badan yang diperlukan adalah:", round(berat, 2), "kg")
```

```python
#Latihan 2.2

x = int(input("Masukan nilai x (bilangan bulat,tidak boleh 0): "))
if x == 0:
    print("Error: x tidak boleh 0")
else:
    hasil = 2 * (x ** 3) + 2 * x + 15/x
    print("Nilai f(x) adalah:", hasil) 
```

```python
#latihan 2.3

gaji_Budi_perjam = float(input("Masukan gaji perjam"))
jam_perminggu = float(input("Masukan jumlah jam kerja perminggu"))
total_jam = jam_perminggu * 5
gaji_kotor_Budi = gaji_Budi_perjam * total_jam
pajak = 0.14 * gaji_kotor_Budi
pendapatan_bersih_Budi = gaji_kotor_Budi - pajak
baju_Budi = 0.10 * pendapatan_bersih_Budi
alat_tulis_Budi = 0.01 * pendapatan_bersih_Budi
sisa = pendapatan_bersih_Budi - baju_Budi - alat_tulis_Budi
sedekah = 0.25 * sisa
anak_yatim = 0.30 * sedekah
kaum_dhuafa = .70 * sedekah

print("\n=== HASIL PERHITUNGAN===")
print("1. Pendapatan Budi sebelum pajak: Rp", gaji_kotor_Budi)
print("2. pendapatan Budi setelah pajak: Rp", pendapatan_bersih_Budi)
print("3. Baju & aksesoris Budi: Rp", baju_Budi)
print("4. Alat tusil Budi: Rp", alat_tulis_Budi)
print("5. Jumlah sedekah: Rp", sedekah)
print("6. Untuk anak yatim: Rp", anak_yatim)
print("7. Untuk dhuafa: Rp", kaum_dhuafa)
```
